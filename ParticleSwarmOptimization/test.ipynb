Libraries:

```python
import numpy as np
import pso
import matplotlib.pyplot as plt
```

Function which will be optimized:

```python
def rosen(xy):
    return 100 *(xy[1] - xy[0] * xy[0])**2 + (1 - xy[0])**2
```

PSO - bare bones

```python
iterations,position=pso.PSO(16,2,rosen,"bare_bones")

print("Rozwiązanie: ",position[-1]," w ",iterations," iteracjach")
```

Plot of the solution for "bare bones"

```python
arg_list=np.arange(-2,2,0.1)
x,y = np.meshgrid(arg_list, arg_list)
z=rosen((x,y))

fig=plt.subplot()
cs = fig.contour(x, y, z,30)
fig.clabel(cs, inline=1, fontsize=5)
fig.set_title('PSO')

for i in range(len(position)):
    fig.scatter(position[i][0],position[i][1],marker=".",color='r')
```
